```python
temp_data = pd.get_dummies(data, columns=['place', 'built', 'sale'])
temp_data.head()
```

```python
temp_data.info()
```

```python
X = temp_data.iloc[:,1:]
y = temp_data.iloc[:,0]
```

```python
print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

```python
y_train_pre = model1.predict(X_train)
y_test_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
print('RMSE on train data :'+str(np.sqrt(mean_squared_error(y_train, y_train_pre))))
print('R2_SCORE on train data :'+str(r2_score(y_train, y_train_pre))+'\n')
print('RMSE on test data :'+str(np.sqrt(mean_squared_error(y_test, y_test_pred))))
print('R2_SCORE on test data :'+str(r2_score(y_test, y_test_pred)))
```

```python
# adjusted r2_score
# Adj r2 = 1-(1-R2)*(n-1)/(n-p-1)
adj_r2 = 1-(1-r2_score(y_test, y_test_pred))*(y_test.shape[0]-1)/(y_test.shape[0]-X_test.shape[1]-1)
adj_r2
```

```python
r2_score2 = model1.score(X_test, y_test)
r2_score2
```
